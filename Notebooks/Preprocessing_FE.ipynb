```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\HP\DATA_SCIENCE\Projects\Credit Risk Scoring\Data\credit_risk_dataset.csv")

df.head()
```

```python
# Data Validation
df = df.drop_duplicates()

df = df[df["person_age"] >= 18]
df = df[df["person_age"] <= 100]
df = df[df["person_income"] > 0]
df = df[df["loan_amnt"] > 0]
df = df[df["loan_int_rate"] >= 0]

df.shape
```

```python
df.isnull().sum()
```

```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df.select_dtypes(include=["object"]).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df["loan_to_income_ratio"] = df["loan_amnt"] / df["person_income"]
```

```python
df["income_band"] = pd.cut(
    df["person_income"],
    bins=[0, 30000, 70000, 150000, float("inf")],
    labels=["Low", "Medium", "High", "Very High"]
)
```

```python
df["age_group"] = pd.cut(
    df["person_age"],
    bins=[18, 25, 35, 50, 100],
    labels=["Young", "Adult", "Middle Age", "Senior"]
)
```

```python
df["high_interest_flag"] = (df["loan_int_rate"] > 15).astype(int)
```

```python
df["high_loan_burden"] = (df["loan_percent_income"] > 0.30).astype(int)
```
