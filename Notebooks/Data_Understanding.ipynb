```python
sns.countplot(data=df, x="loan_status")
plt.title("Loan Status Distribution")
plt.show()
```

```python
# Default Rate by Loan Grade
grade_default = df.groupby("loan_grade")["loan_status"].mean().sort_values()

grade_default.plot(kind="bar")
plt.title("Default Rate by Loan Grade")
plt.ylabel("Default Rate")
plt.show()
```

```python
sns.histplot(df["person_income"], bins=50, kde=True)
plt.title("Income Distribution")
plt.show()
```

```python
sns.histplot(df["loan_amnt"], bins=50, kde=True)
plt.title("Loan Amount Distribution")
plt.show()
```

```python
sns.boxplot(data=df, x="loan_status", y="loan_int_rate")
plt.title("Interest Rate vs Loan Status")
plt.show()
```

```python
intent_default = df.groupby("loan_intent")["loan_status"].mean().sort_values()

intent_default.plot(kind="barh")
plt.title("Default Rate by Loan Intent")
plt.xlabel("Default Rate")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

plt.figure(figsize=(10, 6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
